```python
def outcome(file):
    data = pd.read_csv(file)
    
    train = data['train'].dropna()
    eval = data['eval'].dropna()
    index = [i for i in range(10)]
    df_train = pd.DataFrame(train).reset_index()
    df_eval = pd.DataFrame(eval).reset_index()
    df = pd.concat([df_train, df_eval], axis=1)
    df = df[['train', 'eval']]
    print(df)
    
    fig, ax = plt.subplots()
    ax.plot(df['train'])
    ax.plot(df['eval'])
```
