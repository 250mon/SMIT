```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df=pd.read_excel("../data/PM_ML_Data_FInal.xlsx",index_col="TestID")
```

```python
df.head()
```

```python
sns.pairplot(df,hue='Class',palette='Set1')
```

```python
from sklearn.model_selection import train_test_split
```

```python
seed=7
```

```python
X=df.drop('Class',axis=1)
y=df['Class']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=seed)
```

## k 값 결정하기

```python
n=np.arange(1,40)
```

```python
train_accuracy=np.empty(len(n))
```

```python
test_accuracy=np.empty(len(n))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
for i,k in enumerate(n):
    knn=KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train,y_train)
    train_accuracy[i]=knn.score(X_train,y_train)
    test_accuracy[i]=knn.score(X_test,y_test)
```

```python
plt.plot(n,train_accuracy,label='Training Accuracy')
plt.plot(n,test_accuracy,label='Test Accuracy')
plt.legend()
plt.xlabel('Number of Neighbors')
plt.ylabel('Accuracy')
plt.show()
```

```python
error_rate=[]
```

```python
for i in range(1,40):
    knn_2=KNeighborsClassifier(n_neighbors=i)
    knn_2.fit(X_train,y_train)
    pred_i=knn_2.predict(X_test)
    error_rate.append(np.mean(pred_i !=y_test))
```

```python
plt.plot(range(1,40),error_rate)
plt.xlabel('Number of Neighbors')
plt.ylabel('Error Rate')
```
